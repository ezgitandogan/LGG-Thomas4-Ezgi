In the `data/` folder, you have 10 files that contain Shakespears sonnets. You have to gather all these files into one file `data_all.txt` using threads. Be careful, the sonnets must appear in order.

```python
import os
from threading import Thread, Lock

lock = Lock()

def read_(filename):
    with open(filename, 'r') as file:
        return file.read()

def write_(content):
    with open('data_all.txt', 'a') as file:
        file.write(content)

def gather_():
    files = sorted([f"data/{filename}" for filename in os.listdir("data") if filename.endswith(".txt")])
    threads = []

    for filename in files:
        thread = Thread(target=process_sonnet, args=(filename,))
        thread.start()
        threads.append(thread)

    for thread in threads:
        thread.join()

def process_sonnet(filename):
    sonnet_content = read_(filename)
 
    with lock:
        write_(sonnet_content)



gather_()
```

### Exercise 2
Scrap all the web pages in the `urls` list and display the links. 1 thread per link. 

```python
urls = [
    "http://www.python.org",
    "http://www.python.org/about/",
    "http://www.onlamp.com/pub/a/python/2003/04/17/metaclasses.html",
    "http://www.python.org/doc/",
    "http://www.python.org/download/",
    "http://www.python.org/getit/",
    "http://www.python.org/community/",
    "https://wiki.python.org/moin/",
]
```

```python
import requests
from bs4 import BeautifulSoup
from threading import Thread


urls = [
    "http://www.python.org",
    "http://www.python.org/about/",
    "http://www.onlamp.com/pub/a/python/2003/04/17/metaclasses.html",
    "http://www.python.org/doc/",
    "http://www.python.org/download/",
    "http://www.python.org/getit/",
    "http://www.python.org/community/",
    "https://wiki.python.org/moin/"]

def scrape_url(url):
    
    response = requests.get(url)

   
    if response.status_code == 200:
    
        soup = BeautifulSoup(response.content, 'html.parser')

        links = soup.find_all('a')


        print(f"Links found on {url}:")
        for link in links:
            print(link.get('href'))
    else:
        print(f"Failed to scrape {url}. Status code: {response.status_code}")

def scrape_all_urls():
 
    threads = []
    for url in urls:
        thread = Thread(target=scrape_url, args=(url,))
        thread.start()
        threads.append(thread)

 
    for thread in threads:
        thread.join()

scrape_all_urls()
```
